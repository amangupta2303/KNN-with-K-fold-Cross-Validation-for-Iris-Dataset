```python
from sklearn.datasets import load_iris
```

```python
iris_bunch = load_iris()
```

```python
x = iris_bunch.data
y = iris_bunch.target
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
knn = KNeighborsClassifier(n_neighbors= 12)
```

```python
cross_val_score(knn , x , y , cv=10 , scoring='accuracy').mean()
```

```python
acc_rate = []
for i in range(1,40):
    knn1 = KNeighborsClassifier(n_neighbors= i)
    score1 =cross_val_score(knn1,x,y,cv=10)
    acc_rate.append(score1.mean())
```

```python
error_rate = []
for i in range(1,40):
    knn2 = KNeighborsClassifier(n_neighbors=i)
    score2 = cross_val_score(knn2,x,y,cv=10)
    error_rate.append(1-score2.mean())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(range(1,40),acc_rate,mfc='red',marker='o',linestyle='dashed')
```

```python
plt.plot(range(1,40),error_rate,linestyle='dashed',marker='o',mfc='red')
```

So we can infer that k=13 is the best value for Kneighbor Classifier for the given dataset

```python
knn_best = KNeighborsClassifier(n_neighbors=13)
cross_val_score(knn_best,x,y,cv=10,scoring='accuracy').mean()
```
